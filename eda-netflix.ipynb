## EXPLORATORY DATA ANALYSIS OF FORMATTED DATA

### VISUAL ANALYSIS

```python
missing = missing[missing>0]
missing.sort_values(inplace=True)
plt.figure(figsize=(15,8))
missing.plot.bar()
```

#### OBSERVATION : Most of the values attributed to the "Release Date", "Duration" and "Rating" are Null

### HISTOGRAM

```python
plt.figure(figsize=(12,6))
sns.distplot(a = data["Release Date"], kde = False)
```

```python
sns.kdeplot(data["Release Date"])
```

```python
data["Release Date"].describe()
```

#### OBSERVATION : The following histograms gives the visual representation on the frequency of release date trends. The values are at a minimal increase in the years 1980-2000 and the values peaked in the years 2015-2019 and the trends kept decrementing in the time range 2019-2021 (COVID had an effect in the decrement of movie production and releases)

### Scatter plots

```python
plt.figure(figsize=(12,12))
sns.scatterplot(x = data["Production Country"],y = data["Imdb Score"])
```

#### OBSERVATION : This scatter plot gives the correlation between Production Country and Imdb scores. Argentina is currently leading in the list of Imdb ("The Secret in Their Eyes (2009)").
